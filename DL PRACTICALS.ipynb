Implement a deep-neural feed-forward network for estimating the price of house, given real-estate data
(Boston Housing Price)

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import boston_housing

# Load data
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()

# Build model
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.fit(x_train, y_train, epochs=20, batch_size=32, verbose=1)

# Evaluate
print("Test Loss:", model.evaluate(x_test, y_test))
```

Implement a feed-forward neural networks for classifying movie reviews as positive or negative (using
IMDB dataset)

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import imdb

# Load data
num_words = 10000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=num_words)

# Pad sequences
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=200)
x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=200)

# Model
model = models.Sequential([
    layers.Embedding(num_words, 32),
    layers.Flatten(),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5, batch_size=32)

print("Test Accuracy:", model.evaluate(x_test, y_test))
```

Implement a deep-neural network for classifying news wires by topic (Reuters dataset).

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import reuters

# Load data
num_words = 10000
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=num_words)

# Pad sequences
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=200)
x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=200)

# Model
model = models.Sequential([
    layers.Embedding(num_words, 64),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(46, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=8, batch_size=32)

print("Test Accuracy:", model.evaluate(x_test, y_test))
```

Implement CNN for classifying MNIST dataset


```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape + Normalize
x_train = x_train.reshape(-1, 28, 28, 1) / 255
x_test = x_test.reshape(-1, 28, 28, 1) / 255

# Model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5, batch_size=32)

print("Test Accuracy:", model.evaluate(x_test, y_test))
```


5. Classify MNIST dataset using any pertained model


```python
import tensorflow as tf
from tensorflow.keras import layers, models, datasets, applications

# 1. Load Data
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# 2. Preprocess: reshape, resize, grayscale->RGB, normalize
# Add channel dimension: (60000, 28, 28) -> (60000, 28, 28, 1)
x_train = x_train[..., None]
x_test = x_test[..., None]

# Resize 28x28 -> 32x32
x_train = tf.image.resize(x_train, (32, 32))
x_test = tf.image.resize(x_test, (32, 32))

# Convert grayscale to RGB
x_train = tf.image.grayscale_to_rgb(x_train)
x_test = tf.image.grayscale_to_rgb(x_test)

# Normalize to [0,1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# 3. Load pretrained MobileNetV3Small as feature extractor
base = applications.MobileNetV3Small(
    include_top=False,
    input_shape=(32, 32, 3),
    weights='imagenet',
    minimalistic=True
)
base.trainable = False  # keep pretrained weights frozen

# 4. Add simple classification head
model = models.Sequential([
    base,
    layers.GlobalAveragePooling2D(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Training pretrained MobileNetV3Small on MNIST...")
model.fit(x_train, y_train, epochs=3, batch_size=64, validation_data=(x_test, y_test))
```

Create a model for time-series forecasting using RNN/LSTM


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models

# Generate sample data
x = np.arange(0, 200, 0.1)
y = np.sin(x)

# Prepare sequences
seq_len = 20
X, Y = [], []
for i in range(len(y)-seq_len):
    X.append(y[i:i+seq_len])
    Y.append(y[i+seq_len])

X = np.array(X).reshape(-1, seq_len, 1)
Y = np.array(Y)

# Model
model = models.Sequential([
    layers.LSTM(50, return_sequences=False, input_shape=(seq_len,1)),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, Y, epochs=10, batch_size=32)

# Predict
pred = model.predict(X[:5])
print(pred)
```

Implement an auto-encoder

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# Load data
(x_train, _), (x_test, _) = mnist.load_data()
x_train = x_train.reshape(-1, 28*28) / 255
x_test = x_test.reshape(-1, 28*28) / 255

# Encoder + Decoder
model = models.Sequential([
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(28*28, activation='sigmoid')
])

model.compile(optimizer='adam', loss='mse')
model.fit(x_train, x_train, epochs=5, batch_size=32)

# Reconstruct images
decoded = model.predict(x_test[:5])
print(decoded.shape)
```
